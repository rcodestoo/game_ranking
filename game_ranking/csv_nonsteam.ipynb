```python
df_nonsteam = df[df['SteamStatus'] != 'PC Game (on Steam)']
len(df_nonsteam)
```

```python
df_nonsteam['YouTube ReleaseDate'] = pd.to_datetime(df_nonsteam['YouTube ReleaseDate'], errors='coerce')
```

```python
today = dt.date.today()
```

```python
df_nonsteam['Days_Since_Release'] = (pd.to_datetime(today) - df_nonsteam['YouTube ReleaseDate']).dt.days
```

```python
df_nonsteam
```

```python
df_nonsteam['adjusted_views'] = df_nonsteam['YouTube Views'] / (1 + (df_nonsteam['Days_Since_Release'] / 365))
```
