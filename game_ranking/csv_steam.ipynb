```python
weighted_points=[]
missing_devs = []
print(f'Total rows to process: {len(df_calculation)}')
i = 0
for index, row in df_calculation.iterrows():
    i +=1 
    print(f'Processing row {i} of {len(df_calculation)}')
    if row['Has_Multiple_Developers']:
        # print(f'This row has multiple developers: {row["Developers"]}')
        developers = row['Developers']
        dev_points = []
        for developer in developers:
            # print(f'Searching for developer: {developer} in developer_list')
            weighted_point = developer_list.loc[developer_list['Developer Name'].str.strip().str.lower() == developer.strip().lower(), 'Total Hybrid Weighted Points']
            if not weighted_point.empty:
                dev_points.append(weighted_point.iloc[0].item())
            else:
                dev_points.append(1)
                missing_devs.append(developer.strip())
        avg_weighted_point = sum(dev_points) / len(dev_points) if dev_points else 1
        weighted_points.append(avg_weighted_point)
        df_calculation.loc[index, 'dev_points'] = avg_weighted_point
        # print(f'Found weighted points: {weighted_points}')
    else:
        # print(f'This row has a single developer: {row["Developers"]}')
        developer = row['Developers']
        weighted_point = developer_list.loc[developer_list['Developer Name'].str.strip().str.lower() == developer[0].strip().lower(), 'Total Hybrid Weighted Points']
        if not weighted_point.empty:
            val = weighted_point.iloc[0].item()
        else:
            val = 1
            missing_devs.append(developer[0].strip())
        weighted_points.append(val)
        df_calculation.loc[index, 'dev_points'] = val
    
print(f'Total weighted points appended: {len(weighted_points)}')
print(f'Total missing developers: {len(missing_devs)}')
print(f'Missing developers: {set(missing_devs)}')
print(f'FINAL WEIGHTED POINTS: {weighted_points}')
```

```python
import math

def calculate_follower_weighted_points(followers, min_followers=1000, max_followers=398955, weight=5):
    """
    Calculate weighted points based on follower count using hybrid linear/logarithmic approach.
    Replicates the Excel formula: (0.5*linear + 0.5*log) * weight_factor
    
    Args:
        followers: Number of followers
        min_followers: Minimum threshold (1000)
        max_followers: Maximum threshold (398955)
        weight: Weight multiplier (5 for steam followers)
    """
    
    # Linear normalization to 1-5 scale
    linear_norm = ((followers - min_followers) / (max_followers - min_followers)) * (5 - 1) + 1
    
    # Logarithmic normalization to 1-5 scale
    log_norm = ((math.log(followers) - math.log(min_followers)) / 
                (math.log(max_followers) - math.log(min_followers))) * (5 - 1) + 1
    
    # 50/50 weighted average of linear and logarithmic
    hybrid_score = 0.5 * linear_norm + 0.5 * log_norm
    
    # Multiply by weight factor
    result = hybrid_score * weight
    
    return result

# Calculate follower_points for each game
print(f'Calculating follower weighted points...')
follower_points = []
for index, row in df_calculation.iterrows():
    followers = row['FollowerCount']
    points = calculate_follower_weighted_points(followers, weight=5)
    follower_points.append(points)
    df_calculation.loc[index, 'follower_points'] = points

print(f'Total follower points calculated: {len(follower_points)}')
print(f'Sample follower points: {follower_points[:5]}')
```

```python
# Combine all weighted scores into a final ranking score
# dev_points * 2 + genre_points * 1 + follower_points * 5

print('Calculating final ranking scores...')
for index, row in df_calculation.iterrows():
    dev_weight = 2
    # genre_weight = 1
    follower_weight = 5
    
    final_score = (row['dev_points'] * dev_weight + 
                   row['follower_points'] * follower_weight)
    # row['genre_points'] * genre_weight + 
    df_calculation.loc[index, 'final_ranking_score'] = final_score

print('Final ranking scores calculated!')
df_calculation.head(10)
```
